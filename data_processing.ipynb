## Data Processing

Description: Crop Google Earth images into 500px by 500px square images. Save cropped images in 'before' and 'after' directories.

```python
import math
import numpy as np
from PIL import Image
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

### Cropping Images Demo

```python
def cropImage(im):
    (width, height) = im.size
    (centerWidth, centerHeight) = width/2, height/2
    im = im.crop(
        (
            centerWidth - 250,
            centerHeight - 250,
            centerWidth + 250,
            centerHeight + 250
        )
    )
    return im
```

```python
beforeImageFilePath = 'data/before/25.jpg'
afterImageFilePath = 'data/after/25.jpg'
beforeImage = Image.open(beforeImageFilePath)
afterImage = Image.open(afterImageFilePath)

plt.figure(figsize=(12,12))
plt.subplot(1,2,1)
plt.imshow(np.asarray(beforeImage))
plt.subplot(1,2,2)
plt.imshow(np.asarray(afterImage))
```

```python
beforeImageCropped = cropImage(beforeImage)
afterImageCropped = cropImage(afterImage)

plt.figure(figsize=(12,12))
plt.subplot(1,2,1)
plt.imshow(np.asarray(beforeImageCropped))
plt.subplot(1,2,2)
plt.imshow(np.asarray(afterImageCropped))
```

### Cropping All Images, Saving to File

```python
beforeImagesDir = 'data/before/'
afterImagesDir = 'data/after/'

imageIDs = range(1,26)
for i in imageIDs:
    beforeImagePath = beforeImagesDir + str(i) + '.jpg'
    afterImagePath = afterImagesDir + str(i) + '.jpg'
    beforeImage = Image.open(beforeImagePath)
    afterImage = Image.open(afterImagePath)
    beforeImageCropped = cropImage(beforeImage)
    afterImageCropped = cropImage(afterImage)
    beforeImageCropped.save(beforeImagePath)
    afterImageCropped.save(afterImagePath)
```
